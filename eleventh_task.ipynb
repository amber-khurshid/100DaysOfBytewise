1. Predicting Diabetes Onset Using Logistic Regression

   Dataset: 

Diabetes Dataset   

Preprocessing Steps:

Handle missing values if any.

Standardize features.

Encode categorical variables if any.

  Task:
  
Implement logistic regression to predict diabetes onset and evaluate the model using accuracy, precision, and recall.


```python
import pandas as pd
df= pd.read_csv('diabetes.csv')
df
```

```python
df.dropna
```

```python
df.isnull().sum()
```

```python
X=df.drop(columns=['Outcome'])
y= df['Outcome']

```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
standardized= scaler.fit_transform(X)
standardized = pd.DataFrame(standardized, columns=X.columns)

standardized
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(standardized, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

regression= LogisticRegression()
regression.fit(X_train, y_train)
```

```python
y_predict=regression.predict(X_test)
print(y_predict)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy= accuracy_score(y_test, y_predict)
precision= precision_score(y_test, y_predict)
recall= recall_score(y_test, y_predict)


print("Accuracy: ",accuracy)
print("Precision : ",precision)
print("REcall : ",recall)
```

4. Classifying Emails as Spam Using Decision Trees

Dataset: 

Spam Email Dataset

Preprocessing Steps:

Handle missing values if any.

Standardize features.

Encode categorical variables if present.

Task: 

Implement a decision tree classifier to classify emails as spam or not and evaluate the model using precision, recall, and F1-score.


```python
import pandas as pd
df= pd.read_csv('emails.csv')
df
```

```python
X= df[['text']].values
y=df['spam']
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder= OneHotEncoder(sparse_output= False)
text = df['text'].values.reshape(-1, 1)
X= encoder.fit_transform(text)

df
```

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

standardized_text= scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(standardized_text, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
```

```python
y_pred= tree.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
precision= precision_score(y_test, y_pred)
recall= recall_score(y_test, y_pred)
f1_score= f1_score(y_test, y_pred)
classification_report= classification_report(y_test, y_pred)
print(f"Precision: ", precision)
print(f"REcall score: ", recall)
print(f"f1 score: ", f1_score)

print(f"Classification report: ", classification_report)
```
