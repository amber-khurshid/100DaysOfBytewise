 Predicting Hospital Readmission Using Logistic Regression

    Dataset:

Hospital Readmission Dataset

    Preprocessing Steps:

Handle missing values (e.g., fill missing values with mode for categorical variables).

Encode categorical variables (e.g., one-hot encoding for hospital type, region, etc.).

Standardize numerical features.

    Task:
    
Implement logistic regression to predict hospital readmission and evaluate the model using precision, recall, and F1-score.


```python
import pandas as pd
df=pd.read_csv('hospital_readmissions.csv')
df
```

```python
df.dropna()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder(sparse_output=False)
medical_specialty = df['medical_specialty'].values.reshape(-1, 1)
df['medical_specialty']=encoder.fit_transform(medical_specialty)

age = df['age'].values.reshape(-1, 1)
df['age']=encoder.fit_transform(age)


diag_1 = df['diag_1'].values.reshape(-1, 1)
df['diag_1']=encoder.fit_transform(diag_1)


diag_2 = df['diag_2'].values.reshape(-1, 1)
df['diag_2']=encoder.fit_transform(diag_2)



diag_3 = df['diag_3'].values.reshape(-1, 1)
df['diag_3']=encoder.fit_transform(diag_3)



A1Ctest = df['A1Ctest'].values.reshape(-1, 1)
df['A1Ctest']=encoder.fit_transform(A1Ctest)

change = df['change'].values.reshape(-1, 1)
df['change']=encoder.fit_transform(change)


diabetes_med = df['diabetes_med'].values.reshape(-1, 1)
df['diabetes_med']=encoder.fit_transform(diabetes_med)



readmitted = df['readmitted'].values.reshape(-1, 1)
df['readmitted']=encoder.fit_transform(readmitted)

glucose_test = df['glucose_test'].values.reshape(-1, 1)
df['glucose_test']=encoder.fit_transform(glucose_test)


df
```

```python
X=df.drop(columns='readmitted')
y=df['readmitted']
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaled= scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
regression= LogisticRegression()
regression.fit(X_train,y_train)
```

```python
y_pred=regression.predict(X_test)
# print('predict: ', y_pred)
y_pred
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision= precision_score(y_test, y_pred)
recall= recall_score(y_test, y_pred)
f1_score= f1_score(y_test, y_pred)

print("Precision: ", precision)
print("Recall Score: ", recall)
print("f1 score: ", f1_score)
```

```python
import pandas as pd


df = pd.read_csv('mnist_train.csv')
df
df2 = pd.read_csv('mnist_test.csv')
df2
```

```python
df2 = pd.read_csv('mnist_test.csv')
df2
```

```python
X_train  = df.drop('label', axis=1)
y_train = df['label']
```

```python
X_test = df2.drop('label', axis=1)
y_test = df2['label']
```

```python
X_test=X_test/255
X_train= X_train/255
```

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)

print('Accuracy Score: ', accuracy)
```

```python
from sklearn.metrics import confusion_matrix

confusionMatrix = confusion_matrix(y_test, y_pred)
confusionMatrix
```

```python
from sklearn.metrics import  ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np

disp = ConfusionMatrixDisplay(confusion_matrix=confusionMatrix, display_labels=np.arange(10))

fig, ax = plt.subplots(figsize=(10, 8))  
disp.plot(cmap='Blues', values_format='d', ax=ax)  


plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')

plt.show()
```
