```python
from sklearn.model_selection import train_test_split
X_test, X_train, y_test, y_train= train_test_split(normalized,y,random_state=42, test_size=0.3)
```

```python

print(X_train.shape)  # Should show the shape of training features
print(X_test.shape)   # Should show the shape of testing features
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)  
print(X_train_scaled.shape)  # Should show the shape of training features
print(X_test_scaled.shape)  
```

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(max_depth=1000)
tree.fit(X_train_scaled, y_train)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree, normalized, y, cv=5, scoring='accuracy')
# print(cross_val_score)
print(f"Accuracy for each fold: {scores}")
print(f"Mean accuracy: {scores.mean():.2f}")
```

```python
y_pred=tree.predict(X_test_scaled).astype(int)
print(y_pred)
# 
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
