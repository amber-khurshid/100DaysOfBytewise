Exercise: Create a simple line plot using Matplotlib to display the trend of a list of values over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("titanic.csv")
print(df)

# df = df.dropna(subset=['Age', 'Fare'])

df= df.dropna()
# print(df)
age_sorted = df.sort_values(by='Age')

x = age_sorted['Age']
y = df['Fare']
plt.figure(figsize= (10,6))

plt.plot(x,y ,marker='o', linestyle='-', color='b', linewidth=2, markersize= 6)


plt.title('Trend of Fare over age')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

Exercise: Plot a bar chart using Matplotlib to show the frequency of different categories in a dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("titanic.csv")
# print(df)


df= df.dropna()
# print(df)

# x = df['Fare']
fareCount = df['Fare'].value_counts()
x= fareCount.index
y = fareCount.values

# y = df['Fare']
plt.figure(figsize= (8,6))

plt.bar(x,y ,  color='b')
# , width=0.9


plt.title('Frequency of fare values')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.show()


# d=df['Pclass'].count()
# f= df['PassengerId'].count()
# print(d)
# print(f)

# pclass_count = df['Pclass'].value_counts()
# x=df['Pclass']
# y= df['PassengerId']
# Extract Pclass values (index) and their frequencies (values)
pclass_count=df['Pclass'].value_counts()
x = pclass_count.index
y = pclass_count.values
plt.figure(figsize= (8,6))

plt.bar(x, y ,  color='b')


plt.title('Frequency of Pclass values')
plt.xlabel('Pclass')
plt.ylabel('Frequency')

plt.show()
```

Exercise: Create a scatter plot using Matplotlib to visualize the relationship between two variables in a dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("titanic.csv")
# print(df)


df= df.dropna()
# print(df)

x = df['Age']
y = df['Fare']
plt.figure(figsize= (10,6))

plt.scatter(x,y , color='b')


plt.title('Relationship btw Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

Exercise: Plot a heatmap using Seaborn to visualize the correlation matrix of a dataset.

```python
import seaborn as sns
dataset = sns.load_dataset('penguins')

# include only columns with numeric data
numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns

correlationMatrix= dataset[numerical_columns].corr()
# print(dataset)
# sns.boxplot(dataset)
# sns.heatmap(data=dataset) 
sns.heatmap(correlationMatrix, annot=False, cmap='coolwarm', square=True)
```

Exercise: Use Matplotlib to create a subplot grid that displays multiple charts in a single figure.


```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset = sns.load_dataset('penguins')
# plt.subplot(dataset)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

#  Scatter plot of flipper length vs body mass
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=dataset, ax=axes[0, 0])
axes[0, 0].set_title('Flipper Length vs Body Mass')

# Box plot of bill length by species
sns.boxplot(x='species', y='bill_length_mm', data=dataset, ax=axes[0, 1])
axes[0, 1].set_title('Bill Length by Species')

# Violin plot of bill depth by species
sns.violinplot(x='species', y='bill_depth_mm', data=dataset, ax=axes[1, 0])
axes[1, 0].set_title('Bill Depth by Species')

#  Histogram of flipper length
sns.histplot(data=dataset, x='flipper_length_mm', kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Histogram of Flipper Length')


# Display the plot
plt.show()
```

Exercise: Customize the appearance of a Seaborn plot by changing the color palette, adding titles, and modifying axis labels.

```python
import seaborn as sns

# Load the penguins dataset
dataset = sns.load_dataset('penguins')
# print(dataset)
custom_palette = sns.color_palette("Set3")
# plt.figure(figsize=(10, 6))
ax= sns.boxplot(x='species', y='body_mass_g', data=dataset, hue='sex', palette=custom_palette)

ax.set_title('Body mass of species')
ax.set_xlabel('Species')
ax.set_ylabel('Body Mass')


```

Exercise: Combine Matplotlib and Seaborn to create a complex visualization, such as overlaying a KDE plot on a histogram.

```python
import seaborn as sns
import matplotlib.pyplot as plt
dataset= sns.load_dataset('diamonds')
# print(dataset)
data= dataset['price']
plt.figure(figsize=(10, 6))

sns.histplot(data, kde=True, bins=10, color='blue')
sns.kdeplot(data, color='black', linewidth=15)

plt.title('Histogram with Overlay KDE Plot', fontsize=15)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
